General analysis

```python
df.drop(['Mjob','Fjob','school','address','famsize','reason','guardian','Pstatus','schoolsup','famsup','paid','romantic','higher','internet','activities','nursery'],axis=1,inplace=True)
df
```

```python
filt=df['age']>19
df[filt]
```

```python
x=df.age
y=df.studytime
plt.style.use("seaborn")
plt.bar(x,y,width=0.1)
plt.xlabel("age")
plt.ylabel("studytime")
plt.show()
```

```python
a=df['age']
b=df['health']
plt.style.use("seaborn")
plt.xlabel("age")
plt.ylabel("health")
plt.bar(a,b,width=0.2)
plt.show()
```

Relationship analysis

```python
correlation=df.corr()
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns)
```

```python
sns.pairplot(df)
```

```python
sns.lineplot(x='failures',y='absences',hue='age',data=df)
```

```python
x=df.goout
y=df.age
sns.lineplot(x,y)
```

Summary

1.There are only a few students interested in participating in the survey.
2.Students from the age group of 15-19 are very passionate about their studies.
3.Health is deteriorating for students from 20 years of age.
4.We can observe how variation of absence to classes causes the failure of a subject in a class  of a student.
5.We can see that as the age is increasing,the students want to go out frequently.
